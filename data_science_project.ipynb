```python
df['DEATH'] = [0 if val == '9999-99-99' else 1 for val in df['DATE_DIED']]
```

```python
df.drop(['DATE_DIED'],inplace=True,axis=1)
```

```python
df.drop(columns=["INTUBED","ICU","MEDICAL_UNIT","PREGNANT"], inplace=True)
```

```python
change_cols = ['PATIENT_TYPE', 'PNEUMONIA', 'DIABETES', 'COPD', 'ASTHMA', 'INMSUPR', 'HIPERTENSION', 'OTHER_DISEASE',
               'CARDIOVASCULAR', 'OBESITY', 'RENAL_CHRONIC', 'TOBACCO']

for col in change_cols:
    df[col] = df[col].apply(lambda x: 0 if x==2 else (np.nan if x!=1 else x))
```

```python
df['USMER'].unique()
```

```python
change_cols2 = ['USMER', 'PATIENT_TYPE']

for col in change_cols2:
    df[col] = df[col].apply(lambda x: 0 if x==2 else x)
```

```python
#df['SEX'] = df['SEX'].replace({1: 'Female', 0: 'Male'})
```

```python
df['SEX'] =df['SEX'].apply(lambda x: 1 if x == 1 else 0)
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(df.corr() , annot = True , cmap = "flare")
```

```python

df.dropna(inplace=True)
```

```python
df['POSITIVE'] = df['CLASIFFICATION_FINAL'].apply(lambda x: 1 if x <= 3 else 0)
```

```python
df.drop(columns=['CLASIFFICATION_FINAL'],inplace=True)
```

```python
for col in df.columns:
    condition = df[(df[col] == 97) | (df[col] == 98) | (df[col]== 99)]
    df.drop(condition.index,inplace=True)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,roc_curve,roc_auc_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
X = df.drop(columns='DEATH', axis=1)
Y = df['DEATH']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train,Y_train)
```

```python
predictions = model.predict(X_test)
```

```python
print(accuracy_score(Y_test,predictions))
```

```python
predictions = model.predict(X_train)
```

```python
print(accuracy_score(Y_train,predictions))
```
